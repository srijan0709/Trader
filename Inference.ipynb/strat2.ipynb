```python
import numpy as np
def get_state(df, current_step):
    return np.array([
        df.iloc[current_step]['close'],
        df.iloc[current_step]['high'],
        df.iloc[current_step]['low'],
        df.iloc[current_step]['trades'],
        df.iloc[current_step]['open'],
        df.iloc[current_step]['volume'],
        df.iloc[current_step]['vwap'],
        df.iloc[current_step]['time'],
        df.iloc[current_step]['MA50'],
        df.iloc[current_step]['RSI'],
        df.iloc[current_step]['MACD'],
        df.iloc[current_step]['BB_upper'],
        df.iloc[current_step]['BB_lower'],
        df.iloc[current_step]['ADX'],
        df.iloc[current_step]['CCI'],
        df.iloc[current_step]['ATR'],
        df.iloc[current_step]['ROC'],
        df.iloc[current_step]['OBV']
    ])
```

```python
num_episodes = 1
from tqdm import tqdm
for episode in range(0,num_episodes):
    # state = env.reset()
    total_reward = 0
    number_trans = 0
    wins =0
    lose_win = 0
    lose_defeat =0
    defeat =0
    consecutive_success = 0
    pbar = tqdm(total =500)
    step =0
    next_step = 0

    while step<500:
    
        state = normalize_new_row_with_mean_std(df_test.iloc[step],norm_param)
      
        with torch.no_grad():
            state = torch.tensor(state, dtype=torch.float32).unsqueeze(0).to(device)
            q_values = policy_net(state)
            action=q_values.argmax().item()

        # action = agent.select_action(state)
        
        # print(action)
        # print(state[0][df_test.columns.get_loc('close')].item())
        if action ==1: ## BUY
            next_state = target_buy_test.iloc[step]
           
            next_state_index = next_state["next_state_index"]
            
    
            if(next_state['action']=="Target Hit"):
                wins +=1
                consecutive_success+=1
                
              
            elif next_state['action']=="Stop Loss Hit":
                defeat +=1
                consecutive_success=0
               
                
            elif next_state['action']=="End of Day":
                if next_state['close'] > state[0][df_test.columns.get_loc('close')].item():
                    lose_win+=1
                    print(f"lose_win_buy % {(next_state['close']-state[0][df_test.columns.get_loc('close')].item())*100/state[0][df_test.columns.get_loc('close')].item()}")
                else :
                    lose_defeat+=1
                    print(f"lose_defeat_buy % {(next_state['close']-state[0][df_test.columns.get_loc('close')].item())*100/state[0][df_test.columns.get_loc('close')].item()}")
                consecutive_success=0
               
            
            number_trans +=1
            next_step = next_state_index+1
        
        if action==2: ## Sell  short
           
            next_state = target_sell_test.iloc[step]
            next_state_index = next_state["next_state_index"]
            
            
            if(next_state['action']=="Target Hit"):
                wins +=1
                consecutive_success+=1
    
               
            elif next_state['action']=="Stop Loss Hit":
                consecutive_success=0
                defeat +=1
              
               
            elif next_state['action']=="End of Day":
                consecutive_success=0
                if next_state['close'] < state[0][df_test.columns.get_loc('close')].item():
                    lose_win+=1
                    print(f"lose_win_sell % {(-next_state['close']+state[0][df_test.columns.get_loc('close')].item())*100/state[0][df_test.columns.get_loc('close')].item()}")
                else :
                    lose_defeat+=1
                    print(f"lose_defeat_sell % {(-next_state['close']+state[0][df_test.columns.get_loc('close')].item())*100/state[0][df_test.columns.get_loc('close')].item()}")
                
            number_trans +=1
            next_step = next_state_index+1
        if action ==0:  
            next_step = step+1
            
        pbar.update(next_step - step)
        step = next_step
    pbar.close()
            
    
       
    print(f'Episode {episode + 1}')
    print(f'Number of transcations: {number_trans}, Wins: {wins}, Lose_win: {lose_win},Lose_defeat:{lose_defeat}, Defeat: {defeat} ')

       
       


        

        
```
